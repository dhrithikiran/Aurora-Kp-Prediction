```python
#converting dataset txt file into csv file and saving in the data folder 

import pandas as pd

# Change filename if different
file_path = "data/OMNI_dataset_20252026.txt"

# OMNI files are space-separated (fixed width or whitespace)
df = pd.read_csv(file_path, delim_whitespace=True, header=None)

# Add column names (VERY IMPORTANT)
df.columns = [
    "Year", "DOY", "Hour",
    "Bz_GSM_nT",
    "Proton_Temp_K",
    "Proton_Density_cm3",
    "Solar_Wind_Speed_kms",
    "Flow_Pressure_nPa",
    "Kp_x10"
]

# Convert Kp*10 to real Kp
df["Kp"] = df["Kp_x10"] / 10

# Save as CSV
df.to_csv("data/OMNI_dataset_20252026.csv", index=False)

print("CSV file created!")
df.head()
```
